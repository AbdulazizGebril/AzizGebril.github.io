```python
import pandas as pd
from functools import reduce
```

```python
df=pd.read_csv("Immigration_2.csv")


df_2020=df[["Destination", "Origin","2020"]]


df_2020= df_2020.drop_duplicates(subset=["Destination", "Origin"])


df_agg_2020 = df_2020.groupby(['Origin']).agg({'2020':sum})
df_agg_2020=df_agg_2020.sort_values("2020", ascending=False)


df_2015=df[["Destination", "Origin","2015"]]

df_2015= df_2015.drop_duplicates(subset=["Destination", "Origin"])


df_agg_2015 = df_2015.groupby(['Origin']).agg({'2015':sum})
df_agg_2015=df_agg_2015.sort_values("2015", ascending=False)


df_2010=df[["Destination", "Origin","2010"]]


df_2010= df_2010.drop_duplicates(subset=["Destination", "Origin"])


df_agg_2010 = df_2010.groupby(['Origin']).agg({'2010':sum})
df_agg_2010=df_agg_2010.sort_values("2010", ascending=False)

df_2005=df[["Destination", "Origin","2005"]]

df_2005= df_2005.drop_duplicates(subset=["Destination", "Origin"])


df_agg_2005 = df_2005.groupby(['Origin']).agg({'2005':sum})
df_agg_2005=df_agg_2005.sort_values("2005", ascending=False)

df_2000=df[["Destination", "Origin","2000"]]


df_2000= df_2000.drop_duplicates(subset=["Destination", "Origin"])

df_agg_2000 = df_2000.groupby(['Origin']).agg({'2000':sum})
df_agg_2000=df_agg_2000.sort_values("2000", ascending=False)

df_1995=df[["Destination", "Origin","1995"]]

df_1995= df_1995.drop_duplicates(subset=["Destination", "Origin"])

df_agg_1995 = df_1995.groupby(['Origin']).agg({'1995':sum})
df_agg_1995=df_agg_1995.sort_values("1995", ascending=False)

df_1990=df[["Destination", "Origin","1990"]]

df_1990= df_1990.drop_duplicates(subset=["Destination", "Origin"])

df_agg_1990 = df_1990.groupby(['Origin']).agg({'1990':sum})
df_agg_1990=df_agg_1990.sort_values("1990", ascending=False)

df_final= reduce(lambda x,y: pd.merge(x,y, on='Origin'), [df_agg_1990, df_agg_1995, df_agg_2000,df_agg_2005,df_agg_2010,df_agg_2015,df_agg_2020])



df_final=df_final.sort_values("2020", ascending=False)
df_final.head(60)
```
